```python
df.drop_duplicates(inplace=True)
```

```python
#Step 1: Remove Unnecessary Columns
cols_to_drop = ['URL', 'Domain', 'Title']  # Not useful for ML
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
# Handling Missing Values

# 1ï¸âƒ£ Drop rows where 'label' is missing
df.dropna(subset=['label'], inplace=True)

# 2ï¸âƒ£ Fill missing values in numeric columns with median
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# 3ï¸âƒ£ Fill missing values in categorical columns (only TLD)
cat_cols = ['TLD']
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler

```

```python
le = LabelEncoder()
df['TLD'] = le.fit_transform(df['TLD'])
```

```python
# Avoid division by zero by adding +1 in denominators
df['SpecialCharDensity'] = df['NoOfOtherSpecialCharsInURL'] / (df['URLLength'] + 1)
df['SubDomainDepth'] = df['NoOfSubDomain'] / (df['DomainLength'] + 1)
df['CodeDensity'] = (df['NoOfJS'] + df['NoOfCSS'] + df['NoOfImage']) / (df['LineOfCode'] + 1)
df['RedirectRate'] = df['NoOfURLRedirect'] / (df['NoOfURLRedirect'] + df['NoOfSelfRedirect'] + 1)
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop('label', axis=1))
df_scaled = pd.DataFrame(scaled_features, columns=df.columns[:-1])
df_scaled['label'] = df['label'].values
```
